### **Analyzing Loss**

This is done by specifically taking the entries where the Profit reported is less than 0. These entries are then copied to a new dataset where we will be analyzing loss based on various factors.

```python
loss = retail[retail['Profit']<0]
loss = loss.sort_values(by='Profit',ascending = False)
loss.sort_values('Profit',ascending=True,inplace=True)
loss
```

```python
loss['Location'] = loss['City'].astype(str)+","+loss['State'].astype(str)
loss
```

## **Exploratory Data Analysis - Loss**

### **Ship Mode**

```python
loss['Ship Mode'].unique()
```

```python
shipmode_loc = loss.groupby(['Location','Ship Mode'])[['Location','Ship Mode','Profit']].sum()
shipmode_loc.sort_values('Profit',ascending=True)[:20].plot.barh()
plt.xlabel('Profit')
plt.title('Profit vs Location,Ship Mode')
```

```python
shipmode_st = loss.groupby(['State','Ship Mode'])[['State','Ship Mode','Profit']].sum()
shipmode_st.sort_values('Profit',ascending=True)[:20].plot.barh()
plt.xlabel('Profit')
plt.title('Profit vs State,Ship Mode')
```
