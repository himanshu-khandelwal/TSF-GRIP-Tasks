## **Importing Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## **Importing Dataset**

```python
data = pd.read_csv('iris.csv')
data
```

```python
data['Exact Class'] = data['Species'].map({'Iris-versicolor':0, 'Iris-setosa':1, 'Iris-virginica':2}) 
data 
```

```python
data.info()
```

```python
data.describe()
```

```python
data.columns
```

### **Separating Dataset to Target Variable and Attributes**

The variable which we predict is called the target variable. The rest of the variables in the dataset which is used to predict is called attributes. Here we are splitting dataset to attributes and target variable.

```python
X = data.iloc[:,:-2]
y = data.iloc[:,-1]
X.drop('Id',axis = 1, inplace =  True)
X
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=22) 
```

### **Implementing Model**

```python
model=DecisionTreeClassifier()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
result = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred}) 
result.tail(5)
```

### **Evaluating the Model**

Now we are going to check how reliable this model is by looking into various metrics

```python
import sklearn.metrics as sm
```

```python
# Model Accuracy
print("Accuracy:",sm.accuracy_score(y_test, y_pred))
print('Mean Absolute Error:', sm.mean_absolute_error(y_test, y_pred)) 
print('Mean Squared Error: ', sm.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error: ',np.sqrt(sm.mean_squared_error(y_test, y_pred)))
```

## **Visualizing Decision Tree Classifier**

```python
from sklearn import tree
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 1, figsize = (4,4), dpi = 1000)
tree.plot_tree(model, feature_names = data.columns[:-2], class_names = data['Species'].unique(), filled = True);
```
